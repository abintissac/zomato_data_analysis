```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("zomato1.csv")
df
```

```python
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis = 1)
```

```python
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

```python
 df['rate'].unique()
```

```python
 def handlerate(value):
        if(value=='NEW' or value=='-'):
            return np.nan
        else:
            value = str(value).split('/')
            value = value[0]
            return float(value)
 df['rate'] = df['rate'].apply(handlerate)
 df['rate'].head()
```

```python
df['rate'].unique()
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
```

```python
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df.drop(['city'],axis=1)
```

```python
df['cost'].unique()
```

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['cost'] = df['cost'].apply(handlecomma)
df['cost'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
location = df['location'].value_counts(ascending = False)
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location_less300(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location_less300)
df['location'].value_counts()
```

```python
df.head(25)
```

```python
df['cuisines']
```

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines < 100]
def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
        
```
